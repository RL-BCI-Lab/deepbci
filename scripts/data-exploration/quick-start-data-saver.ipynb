```python
def extract_info(data: dbci.data_utils.data.DataGroup, exclude=['tags', 'metadata']):
    info_dict = {}
    for key, value in data.__dict__.items():
        if key not in exclude:
            info_dict[key] = value
    return info_dict

def grps_to_dict(grps, exclude=['tags', 'metadata']):
    data_dict = grps.data_map.to_dict()
    
    for key, value in data_dict.items():
        info_dict = extract_info(value, exclude=exclude)
        data_dict[key] = info_dict
        
    return data_dict
```

```python
data = grps_to_dict(grps)
data
```

```python
def pickle_data(filepath, data):
    with open(filepath, 'wb') as f:
        pickle.dump(data, f)
```

```python
data_dir = 'quick-start-data'
if not os.path.exists('quick-start-data'):
    os.makedirs('quick-start-data')
filepath = join(data_dir, 'S01-all-dict.pickle')
pickle_data(filepath, data)
```
